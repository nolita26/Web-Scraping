```python
import requests
from bs4 import BeautifulSoup
```

```python
#Working with BeautifulSoup Library
response = requests.get("https://nextjs.org/docs")
bs = BeautifulSoup(response.text,"lxml")
print(bs.find("p").text)
```

```python
#Working with Requests Library
url_to_parse = "https://nextjs.org/docs"
response = requests.get(url_to_parse)
print(response)
# print(response.text)
```

```python
#Create an object of BeautifulSoup
soup = BeautifulSoup(response.text,"lxml")
print(soup.title)
```

```python
#To get the text inside the element, use the text attribute.
print(soup.title.text)

#Return the first h1 tag it finds
print(soup.h1)
```

```python
#Used to find elements based on name, attributes, or text. This should cover most of the scenarios. For scenarios like finding by class, there is **kwargs that can take other filters.
soup.find(name=None, attrs={}, recursive=True, text=None) #**kwargs
```

```python
#Returns the first div it finds
soup.find("div",id="toc")
```

```python
soup.find("div",class_="toc") #not the underscore
soup.find("div",{"class": "toc"})  #Use a dictionary as the second argument
soup.find("div",{"class": "toc", "id":"toc"}) #By using a dictionary, more than one attribute can be specified
```

```python
soup.find_all("span",class_="toctext")
soup.find_all("span",class_="tocnumber") #heading numbers are extracted
```

```python
table_of_contents = soup.find("div",id="toc") #Finding Nested Elements
data = []
if table_of_contents is not None:
    # find_all method will return a list of elements, so we can iterate over it
    for heading in table_of_contents.find_all("li"):
        heading_text = heading.find("span", class_="toctext").text
        heading_number = heading.find("span", class_="tocnumber").text
        data.append({
            'heading_number': heading_number,
            'heading_text': heading_text,
        })
else:
    print("Table of contents not found.")
```

```python
file= open("toc.csv", "w", newline="") #explore the data
```

```python
import csv
#create an instance of DictWriter object. This needs a list of headers. In this case, this is going to be the dictionary keys in the data.
writer = csv.DictWriter(file,fieldnames=['heading_number','heading_text'])
```

```python
writer.writeheader()
writer.writerows(data)	#method to write the data
```
